# traversal in df

```python
df = pd.DataFrame([[1,2,3],[4,5,8],[34,56,78]],columns = ['col1','col2','col3'])  
df
```

```python
for i in df.iterrows():
    print(i)
    print("\n")
```

```python
for i in df.iterrows():
    print("row number  ",i[0])
    print("data\n",i[1].values)
    print("\n")
```
