# Merging is used when we need to join two dataframes having different columns and there is a column which is common in dataframes

### It combines the dataframe columns wise only

### while concat combines the dataframes column wise and rows wise. 
There are four basic ways to handle the merge 



1-inner (default)


2-left


3-right

4-outer

#  MERGE METHOD	JOIN NAME	DESCRIPTION
left	 	Use keys from left frame only


right		Use keys from right frame only


outer		Use union of keys from both frames


inner	     Use intersection of keys from both frames

```python
import pandas as pd
df1 = pd.DataFrame({'key': ['K0', 'K8', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="right")
```

```python
pd.merge(df2,df1,on="key",how="left")
```

```python
pd.merge(df2,df1,on="key",how="right")
```

```python
pd.merge(df1,df2,on="key",how="outer")
```

```python
pd.merge(df2,df1,on="key",how="outer")
```

# merging when all column names are same

```python
import pandas as pd
df1 = pd.DataFrame({'key': ['K0', 'K8', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'A': ['C0', 'C1', 'C2', 'C3'],
                          'B': ['D0', 'D1', 'D2', 'D3']})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how="left",on="key")
```

```python
pd.merge(df1,df2,how="right",on="key")
```

```python
pd.merge(df1,df2,how="inner",on="key")
```

```python
pd.merge(df1,df2,how="outer",on="key")
```

# merging on more than one  common column

```python
df1 = pd.DataFrame({'key1': ['K0', 'K8', 'K2', 'K3'],
                    'key2': ['K10', 'K18', 'K12', 'K13'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
df2 = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
                    'key2': ['K10', 'K11', 'K12', 'K13'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how="left",on=["key1","key2"])
```

```python
pd.merge(df1,df2,how="right",on=["key1","key2"])
```

```python
pd.merge(df1,df2,how="inner",on=["key1","key2"])
```

```python
pd.merge(df1,df2,how="outer",on=["key1","key2"])
```
